# Basic Plot (Line Plot)
ðŸ“Œ Definition

A line plot shows data points connected by straight lines.
Used to show trends over time.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [20,30,60,70,90]
plt.plot(x,y)
plt.show()
```

 # Adding Labels & Title :

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
# line-style (ls)  ,line-width (lw) , line-color(c)


import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,10,2)
y = x**2
print(x,y)
plt.plot(x , y , ls = "solid",c = "orange" , lw = "4")
plt.plot(x,y,marker = "o")
plt.show()
```

 # Front Properties

```python
# 1. Arial ,2.cambria

x = np.arange(0,20,2)

y = x*10
plt.plot(x,y)

f1 = { "family" : " cambria", 'color' : 'red', 'size' : 15}
plt.title("Front",fontdict = f1)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
a=np.arange(0,10,2)
b=[i**2 for i in a]
print(a,b)
plt.plot(a,b , marker="o",mec='r',mfc='b',ms=10,ls='dashed',c='g',lw='3')
f1={"family":"Arial","size":"20","color":"r"}
plt.title("these is plot",fontdict=f1)
f2={"family":"calibri","size":"20","color":"b"}
f3={"family":"cambria","size":"20","color":"w"}
plt.plot(a,b,label="cube")
plt.plot(a,b,label="square")

plt.xlabel("x_label",fontdict=f2)
plt.ylabel("y_label",fontdict=f3)
plt.grid(axis="both",ls="solid",lw=10,c='c')
plt.legend(loc="upper right",framealpha=1,facecolor="yellow",shadow=True,fancybox=True)
plt.show()
```

# Subplots (Multiple plots in one figure) :

```python
x = [1,2,3,4,5]
y1 = [10,20,30,40,50]
y2 = [30,40,50,70,80]
y3 = [40,50,60,70,80]
y4 = [60,70,80,90,100]
plt.suptitle("MULTIPLE PLOTS")

plt.subplot(2,2,1) # 2-rows ,2-columns , 1-plot
plt.plot(x,y1)
plt.title("PLOT-1")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("PLOT-2")

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("PLOT-3")

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("PLOT-4")

plt.show()
```

 # LEGEND FUNCTIONS :

```python
# 1.loc (Location)
# 2.frame-alpha = 0 or 1
# 3.frame-color 
# 4.edge-color 
# 5.shadow = True or False
# 6.fancy-box = True or False

x =np.arange(0,10,2)
y1 = x**2
y2 = x**3

plt.plot(x,y1 ,label ="square")
plt.plot(x,y2 ,label ="cubes")

plt.legend(loc = "upper left" , framealpha = 1 , facecolor = 'yellow' , edgecolor = 'blue' , fancybox = True ,shadow =True ,fontsize = 17)

plt.show()
```

# CMAP AND COLORBAR :

# Pie Chart
ðŸ“Œ Definition

Shows percentage or proportion.

ðŸ“ Example:

Market share

Budget usage


```python
# parameters :
# 1 . Start angle = degrees like 0 , 180, 60
# 2. Explode = it divide the charts into separate with values
# 3. autopct = means auto percentage ,percentage show ,start with " % (value) % "

from matplotlib import pyplot as plt
student = ('excellent','good','avg','bad')
values = (12,8,22,30)

plt.figure(figsize=(8,10))

plt.pie(values , labels = student , startangle = 90 , explode = (0,2,0,0) , shadow = True , colors = ('green','red','blue','yellow') ,autopct ="%2.1f%%")
plt.legend(title="performance")

plt.show()
```

```python
from matplotlib import pyplot as plt
student=["exe","good","avg","bad"]
values=[23,45,67,89]
plt.figure(figsize=(8,10))
plt.pie(values, labels = student, startangle = 90,explode = [0,0,0,0] ,shadow = True, colors = ["red","pink","blue","black"], autopct = "%2.1f%%")
plt.legend(title="performance")
plt.show()
```

 # Histogram
ðŸ“Œ Definition

Shows frequency distribution of numerical data.

ðŸ“ Example Uses:

Exam score distribution

Age distribution

```python
# hist types [ 1.bar , 2.step , 3.stepfilled ]

marks = [23,34,56,78]
grade = [0,32,45,67]
plt.title("student")

plt.hist(marks , grade , histtype = "step" , rwidth = 0.7 , facecolor = "blue") 

plt.xticks([0,32,45,67])
plt.xlabel("dd")
plt.ylabel("kdmk")

plt.show()
```

```python
# xlim
# ylim

x = (18,23,44,54,55)
y = (34,45,67,87,89)
plt.plot(x,y)

plt.xlim([10,100])
plt.ylim([50,90])

plt.show()
```
