ARITHMETIC OPERATIONS : 


```python
import numpy as np
a=np.arange(1,100,10)

b=np.arange(2,100,10)
print(a,b)
```

```python
# 1.Addition
z=np.add(a,b)

# 2.Subtract
y=np.subtract(a,b)

# 3.multiplication
x=np.multiply(a,b)

# 4.Division
w=np.divide(a,b)

# 5.Exponant
#v =np.exp(a,b)

#6.Square root
#u =np.sqrt(a,b)

# 7.Equal
t=np.array_equal(a,b)

```

```python
print(f'Addition of a,b :{z}')
print(f'Subtraction of a,b :{y}')
print(f'multiplication of a,b :{x}')
print(f'Division of a,b :{w}')
#print(f'EXP of a,b :{v}')
#print(f'Square root  of a,b :{u}')
print(f'Equal of a,b :{t}')
```

ARRAY FUNCTIONS:

```python
a=np.arange(10,100,10)

# 1.minimum 
q=np.min(a)

# 2.Maximum
w=np.max(a)

# 3.Sum
e=np.sum(a)

# 4.Mean
r=np.mean(a)

# 5.Median
t=np.median(a)

# 6.Variance

y=np.var(a)

# 7.Standard Diviation

u=np.std(a)
```

```python
print(f'The Array Elements are :{a}')
print(f'Minimum Number is :{q}')
print(f'Maximum Number is :{w}')
print(f'Sum Number is :{e}')
print(f'Mean Number is :{r}')
print(f'Median Number is :{t}')
print(f'Variance Number is :{y}')
print(f'Std Number is :{u}')
```
