# 8️⃣ Broadcasting :
**Broadcasting** allows NumPy to perform **operations on arrays of different shapes automatically.**

Normally, arrays must have the same shape to perform operations.
But with broadcasting, NumPy intelligently **expands the smaller array to match the larger one.**

This feature is heavily used in **Scikit-learn, TensorFlow, and PyTorch.**

**Broadcasting** is used heavily in **machine learning computations** such as:

- Feature scaling
- Gradient calculations
- Neural network weight updates

# 1 — What is Broadcasting? 
Broadcasting allows operations like:

```
array + number
array + smaller_array
```


NumPy automatically expands the **smaller value to match the shape.**

Example:

[10 20 30] + 5

Result:

[15 25 35]

The number 5 is broadcast to all elements.

# 2 — Broadcasting with Scalars


````python
import numpy as np
arr = np.array([10,20,30])
result = arr+ 5
print(result)
````

# 3 — Broadcasting Between Arrays

Broadcasting also works between arrays of different shapes.

Example:

[1 2 3]
+
[10]

Result:

[11 12 13]
The smaller array is expanded automatically.

# 4 — Broadcasting with 2D Arrays


````python
import numpy as np 
arr =np.array ([
    [1,2,3],
    [4,5,6]
])
result = arr + 10
print(result)
````

# 5 — Broadcasting with Different Shapes

````python
import numpy as np
arr1 = np.array([
    [1,2,3,4],
    [5,6,7,8]
])
arr2 = np.array([10,20,30,40])
result = arr1 + arr2
print(result)
````

# 🟦 Broadcasting Rules

Broadcasting works when these rules are satisfied.

**Rule 1**

Dimensions must be **equal OR one of them must be 1.**

**Rule 2**

Comparison happens from **right to left dimensions.**

Example:

(2,3)
(3,)

**Rightmost dimensions match → broadcasting works.**

Example of Invalid Broadcasting

- (2,3) -> 2 rows 3 columns

- (2,2) ->2 rows 2 columns

This causes an **error because dimensions don't match.**

````python
# pratice

import numpy as np

A = np.array([
    [10,20,30],
    [40,50,60]
])

B = np.array([1,2,3])

print("Addition :" ,A+B)
print("Multiplication :",A*B)
````

# 🔟 Axis Operations (axis = 0 , axis = 1) 
In NumPy, axis tells Python the direction of operation.

When we perform operations like:

- sum()
- mean()
- min()
- max()

NumPy needs to know **whether to operate across rows or columns.**

So we use axis.

**Rule:**
|Axis	 | Meaning                                  |
|--------|------------------------------------------|
|axis = 0|	Operation down the rows (column-wise)   |
|axis = 1|	Operation across the columns (row-wise) |

````python
import numpy as np

arr = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

print("Array:")
print(arr)

# axis = 0
print("\nSum axis=0 (column wise):")
print(np.sum(arr, axis=0))

# axis = 1
print("\nSum axis=1 (row wise):")
print(np.sum(arr, axis=1))
````

````python
# pratice :

import numpy as np
arr = np.array([
    [5,10,15],
    [20,25,30],
    [35,40,45]
])
print("\nArray:",arr)

print("\n 1️⃣ sum using axis=0:",np.sum(arr, axis = 0))
print("\n 2️⃣ sum using axis=1:",np.sum(arr , axis = 1))
print("\n 3️⃣ Find mean using axis=0 :",np.mean(arr , axis = 0))
print("\n 4️⃣ Find max using axis=1:",np.max(arr , axis = 1))
````

# 2️⃣ flatten() — Convert to 1D Array

***flatten()*** converts a **multi-dimensional array into a 1D array.**

It also creates a **copy of the array.**

**Important Point:**

flatten() does **NOT modify the original array.**

````python
import numpy as np

arr = np.array([
    [1,2,3],
    [4,5,6]
])

flat = arr.flatten()

print(flat)
````

# 3️⃣ ravel() — Flatten Without Copy


ravel() also **converts array to 1D, but it returns a view instead of a copy.**

This means **it uses the same memory.**

````python
import numpy as np

arr = np.array([
    [1,2,3],
    [4,5,6]
])

flat = arr.ravel()

print(flat)
````

# Key Difference :
|Function	| Copy	| Memory Usage|
|-----------|-------|-------------|
|flatten()	|Yes	|More         |
|ravel()	|No	    |Less         |

In large datasets, **ravel() is faster and more memory efficient.**

# 12️⃣ Joining Arrays :

Joining arrays means **combining two or more arrays into one larger array.**

NumPy provides several functions:

- concatenate()
- vstack()
- hstack()

These are used frequently when **merging features or datasets in Machine Learning.**

# 1️⃣ concatenate() — General Joining


np.concatenate() **joins arrays along an existing axis.**

````python
import numpy as np

arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

result = np.concatenate((arr1, arr2))

print(result)
````

# concatenate() with 2D Arrays


````python
arr1 = np.array([
    [1,2],
    [3,4]
])

arr2 = np.array([
    [5,6],
    [7,8]
])

result = np.concatenate((arr1, arr2), axis=0)

print(result)
````

# 3️⃣ vstack() — Vertical Stack:

vstack() stacks **arrays row-wise (top to bottom).**

**It is equivalent to:**

**concatenate(axis=0)**

````python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

result = np.vstack((arr1, arr2))

print(result)
````

# 4️⃣ hstack() — Horizontal Stack


hstack() stacks **arrays column-wise (side by side).**

**Equivalent to:**

concatenate(axis=1)

````python
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

result = np.hstack((arr1, arr2))

print(result)
````

````python
# 2D Example
arr1 = np.array([
    [1,2],
    [3,4]
])

arr2 = np.array([
    [5,6],
    [7,8]
])

np.hstack((arr1,arr2))
````

````python
# pratice
import numpy as np

A = np.array([10,20,30])
B = np.array([40,50,60])

print("Concatenate:",np.concatenate((A,B)))
print("Vstack:",np.vstack((A,B)))
print("Hstack:",np.hstack((A,B)))
````

### 🧠 Why Double Parentheses?

Because we pass multiple arrays as one argument (tuple).

Example:

np.concatenate((A,B))

This means:

### 21️⃣ Vectorization :
Vectorization means **performing operations on entire arrays at once instead of using loops.**

Instead of processing **one element at a time**, NumPy processes **all elements together.**

# 3️⃣ Code Example
### ❌ Without Vectorization (Using Loop)

````python
import numpy as np

arr = np.array([1,2,3,4,5])

result = []

for i in arr:
    result.append(i * 2)

print(result)
````

````python
import numpy as np

arr = np.array([1,2,3,4,5])

result = arr * 2

print(result)
````

# 4️⃣ Code Explanation
## Loop Version
Python processes **element by element
Slow for large datasets**

## Vectorized Version
NumPy processes the **whole array in one operation
Very fast**

This happens because NumPy is implemented in C internally, which is faster than Python loops.

# Performance Example

````python
import numpy as np
import time

arr = np.arange(1000000)

# Loop method
start = time.time()

result = []
for i in arr:
    result.append(i * 2)

end = time.time()
print("Loop time:", end-start)
````

## Vectorized version:

````python
start = time.time()

result = arr * 2

end = time.time()
print("Vectorized time:", end-start) # Vectorized code is much faster
````

# Feature Scaling (Very Important)

Machine learning models work better when **features are on similar scales**.

#### Example dataset:
```
Age = 20,30,40
Salary = 20000,80000,150000
```
Salary values are much larger.

We scale them using normalization.

````python
# Normalization
import numpy as np

data = np.array([10,20,30,40,50])
 
normalized = (data - np.min(data)) / (np.max(data) - np.min(data)) # formula of normalization

print(normalized)
````

# 28️⃣ Feature Scaling using NumPy


Feature scaling means adjusting values of different features so that they are in a similar range.

#### Example dataset:
```
Age        → 20, 30, 40
Salary     → 20000, 80000, 150000
```
##### Problem:

- Salary values are much larger than Age values

- This can confuse machine learning models.

- Feature scaling fixes this issue.

````python
# 3️⃣ Code Example — Min-Max Scaling

import numpy as np


data = np.array([10,20,30,40,50])

scaled = (data - np.min(data)) / (np.max(data) - np.min(data))

print(scaled)
````

````python
# Example with Dataset
import numpy as np

data = np.array([
    [25,50000],
    [30,60000],
    [35,80000],
    [40,100000]
])

scaled = (data - data.min(axis=0)) / (data.max(axis=0) - data.min(axis=0))

print(scaled)
````

# Another Important Scaling — Standardization

````python
import numpy as np

data = np.array([10,20,30,40,50])

standardized = (data - np.mean(data)) / np.std(data)

print(standardized)
````
