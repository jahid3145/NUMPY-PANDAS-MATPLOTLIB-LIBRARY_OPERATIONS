# loc and iloc Functions :

In Pandas, loc and iloc are the primary tools used to select specific rows and columns from a DataFrame.

The easiest way to remember the difference is:

loc is label-based (uses names).

iloc is integer-based (uses index numbers).
 

# 1. loc (Label-based Selection)
Definition: loc stands for "location." it allows you to access a group of rows and columns by their labels (the names you see at the top of columns or the side of rows).

Syntax: df.loc[row_label, column_label]

```python
# loc operations  >>> ROWS ARE DEFINE AS NUMBERS AND COLUMNS ARE DEFINED AS NAMES
# 1. Show the first row 
z = df.loc[1]

# 2. show row number + column name

x = df.loc[5,['hotel']]

# 3. slicing shows 10 rows

y = df.loc[0:10]

# 4. slicing + column name

v = df.loc[0:10 , "hotel"]

# 5.show multiple  slicing + column name

t = df.loc[0:10 , ['lead_time' , 'arrival_date_year']]

# 6. shows slicing + between columns using :(colon)

r = df.loc[0:10,'hotel' : 'arrival_date_year']
```

```python
print(f'\n 1.shows the one row : \n{z}\n')
print(f'\n 2.shows the  row + column name : \n{x}\n')
print(f'\n 3.slicing shows 10 rows : \n{y}\n')
print(f'\n 4. slicing + column name : \n{z}\n')
print(f'\n 5.show multiple  slicing + column name : \n{z}\n')
print(f'\n 6. shows slicing + between columns using :(colon): \n{z}\n')

```

# 2. iloc (Integer-based Selection)
Definition: iloc stands for "integer location." It allows you to access data by its positional index (0, 1, 2, etc.), just like you would with a standard Python list.

Syntax: df.iloc[row_index, column_index]

```python
# iloc operations >> ROWS ARE DEFINED AS NUMBERS AND COLUMNS ARE DEFINED AS NUMBERS

# 1 SHOWS ONE ROW AND COLUMN 

m = df.iloc[1,4]  # it  shows fist row of 4th column

# 2. slicing

# iloc[row-start : row stop , col_start : col_stop]

n = df.iloc[0:4,0:4] # it shows 1st 0-4 rows records and 1st 0-4 column records


# 3. slicing +coulumn number

l = df.iloc[0:4,1] # it shows 1st col of 0-4 rows

# 4. shows the multiple rows 
k = df.iloc[[1,2,3,4]]

# 5. shows the multiple columns 

j = df.iloc[ : ,[2,3]] # it shows 2,3 columns

# 6.slicing + columns

h = df.iloc[0:10 ,[1,2,3,4]]
```

```python
print(f'\n 1. SHOWS ONE ROW AND COLUMN \n It  shows fist row of 4th column :\n{m}\n')
print(f'\n 2. slicing \n It shows 1st 0-4 rows records and 1st 0-4 column records :\n{n}\n')
print(f'\n 3. slicing +coulumn number \n It shows 1st col of 0-4 rows : \n {l}\n')
print(f'\n 4. shows the multiple rows : \n{k}\n')
print(f'\n 5. shows the multiple columns \n  It shows 2,3 columns :\n {j}\n')
print(f'\n  6.6.slicing + columns : \n {h} \n ')
```

# DATA FILTERING AND CONDITIONAL CHANGES

1. Simple Data Filtering
Filtering is usually done by passing a Boolean Condition (a True/False test) inside the square brackets [].

2. 2. Multiple Conditions
To filter using more than one rule, use the "Bitwise Operators":

& (AND): Both conditions must be True.

| (OR): At least one condition must be True.

Note: You must wrap each condition in parentheses ().

3. Conditional Operations (Updating Data)
Sometimes you don't just want to see the data; you want to change it based on a condition. For this, we use .loc.

Pattern: df.loc[row_condition, column_to_change] = new_value

4.5. Filtering String Data
You can filter based on parts of a word using .str.contains().




```python
#  1 . less or greater than

qq = df.loc[df["adults"]<10]

# 2. with and condition

ww = df.loc[(df["adults"]<10) & (df["adults"]>10)]

# 3.STRING CONTAINS

tt = df.loc[df["name"].str.contains("i")]

# 4. Start with letter

ff = df.loc[df["name"].str.startswith("a")]

# 5. End With

gg = df.loc[df["name"].str.endswith("t")]
```

```python
print(f' \n  1 . less or greater than:\n{qq}\n')
print(f'\n 2. with and condition :\n {ww}\n')
print(f'\n 3.STRING CONTAINS :\n {tt}\n')
print(f'\n 4. Start with letter :\n {ff}\n')
print(f'\n 5. ends with letter :\n {gg}\n')

```

 # CHANGING MULTIPLE CONDITIONS : 
  

1 Using .loc for Multiple Conditions
This is the best method if you want to update values in an existing column or create a new one based on several rules joined by "AND" (&) or "OR" (|).

Rules to remember:

Each condition must be inside parentheses ().

& = All conditions must be true.

| = At least one condition must be true.

```python
# Example: Create a 'Category' for bookings
# If it's a Resort Hotel AND lead_time is over 200 days
hh = df.loc[(df['hotel'] == 'Resort Hotel') & (df['lead_time'] > 200), 'Category'] = 'Long-term Resort'

# If it's a City Hotel OR the price is very high
vv = df.loc[(df['hotel'] == 'City Hotel') | (df['adr'] > 500), 'Priority'] = 'High Attention'
```

```python
print(hh)
print(vv)
```

2. Using np.select (The "Decision Tree" Approach)
If you have 3 or more different conditions and you want a different result for each, np.select is much cleaner than writing multiple .loc lines. It works like a "Switch" or "Case" statement.

```python
import numpy as np

# 1. Define your list of conditions
conditions = [
    (df['lead_time'] < 7),                          # Condition 1: Last minute
    (df['lead_time'] >= 7) & (df['lead_time'] < 30), # Condition 2: Short term
    (df['lead_time'] >= 30)                         # Condition 3: Planned
]

# 2. Define the values you want for each condition
choices = ['Immediate', 'Upcoming', 'Advanced']

# 3. Apply it
bb = df['Booking_Type'] = np.select(conditions, choices, default='Unknown')
```

```python
print(bb)
```

3. Using .mask() and .where()
These are specialized functions for swapping values based on conditions.

.where(): Keeps the value if the condition is True, otherwise replaces it.

.mask(): Replaces the value if the condition is True.

```python
# If the lead_time is less than 0, change it to 0 (useful for cleaning data)
nn = df['lead_time'] = df['lead_time'].mask(df['lead_time'] < 0, 0)
```
