# Project 1: Student Marks Analyzer (NumPy)
### Project Goal :

Analyze student marks using NumPy and extract useful insights such as:

- Average marks
- Highest marks
- Students who passed
- Students who scored above average

```python
import numpy as np

#Student marks
marks = np.array([45,67,89,32,76,51,90,40,66,72])

#Average Marks
average = np.mean(marks)
print("Average Marks:",average)

# Highest Marks

highest = np.max(marks)
print("Highest Marks :",highest)

# Passed Students

passed = marks[marks >= 40]
print("Passed Students :",passed)

#Failed Student

failed = marks[marks <40]
print("Failed Student:",failed)

#Above Average Srudents

above_avg = marks[marks > average]

print("above Aveage:",above_avg)

# Lowest marks
lowest = np.min(marks)
print("Lowest Marks :", lowest)

# Count Total  Passed Students
print("Total  Passed Students:",len(passed))
```

# Project: Student Performance Analyzer
### Project Goal

Analyze student performance across multiple subjects and extract insights such as:

- Student average marks
- Class topper
- Subject-wise average
- Students above class average

```python
import numpy as np

students = np.array(["Ali","Sara","John","Ravi","Meana"])

marks = np.array([
    [85,78,92,88],
    [76,85,80,79],
    [90,91,89,93],
    [60,65,70,68],
    [88,84,86,90]
])

# Total Marks
total_marks = np.sum(marks,axis=1)
print("Total Marks:",total_marks)

#Average marks per student
average_marks = np.mean(marks,axis=1)
print("Average Marks:",average_marks)

#class topper
topper_index = np.argmax(total_marks)
topper_name = students[topper_index]
print("Class Topper:",topper_name)

#Subject Average

subject_avg = np.mean(marks, axis=0)
print("Subject Average:",subject_avg)

# Class average
class_avg = np.mean(total_marks)

# Above Aveage Student
above_avg_students = students[total_marks > class_avg]
print("Student Above Class Aveage:",above_avg_students)
```

# Project -2 **Sales Data Analyzer**
### 🎯 Project Goal:

Analyze company sales data to answer real business questions:

- Which product sells the most
- Which month has highest revenue
- Which region performs best
- What is the average daily sales
- Identify top performing products

```python
import numpy as np

# Create Sales Dataset

products = np.array(["Laptop", "Phone", "Tablet", "Headphones"])

sales = np.array([
    [200, 350, 150, 400],
    [220, 370, 160, 420],
    [250, 390, 180, 450],
    [300, 420, 200, 480],
    [320, 450, 210, 500],
    [350, 480, 230, 520],
    [370, 500, 240, 540],
    [400, 520, 260, 560],
    [420, 550, 270, 580],
    [450, 580, 300, 600],
    [480, 600, 320, 620],
    [500, 650, 350, 700]
])

print("shape:",sales.shape)

# Total Sales Per Product

total_product_sales = np.sum(sales , axis =0)
print("Total Sales Per Product:",total_product_sales)

# Monthly Total Sales
monthly_sales = np.sum(sales, axis=1)
print("Monthly Total Sales:",monthly_sales)

# Best Selling Product
best_product_index = np.argmax(total_product_sales) #It returns the index (position) of the largest value, not the value itself.
best_product = products[best_product_index]
print("Best Selling Product:",best_product)


# Highest Sales Month
best_month = np.argmax(monthly_sales)
print("Highest Sales Month:",best_month)

# Average Monthly Sales
avg_sales = np.mean(sales)
print("Average Monthly Sales:",avg_sales)

# Sales Variability
sales_std = np.std(monthly_sales)
print("sales Standard Deviation :",sales_std)

#Reshaping Sales Data
#Suppose we want quarterly sales.
#Each quarter = 3 months

quarterly_sales = sales.reshape(4,3,4)
#(4 quarters, 3 months, 4 products)
quarter_total = np.sum(quarterly_sales, axis=1)

print("Quarterly Sales:\n", quarter_total)

# Top 2 Products
top_products_index = np.argsort(total_product_sales)[-2:] #[-2:] means take the last 2 elements.
# argsort() returns indices that would sort the array.
#Sales sorted ascending

top_products = products[top_products_index]

print("Top 2 Products:", top_products)
```

# Project -3 **Image Processing with NumPy**
###  Project Goal

Perform basic image processing operations using NumPy:

- Load an image as an array
- Analyze pixel values
- Crop an image
- Adjust brightness
- Convert to grayscale
- Apply simple filters

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a Simulated Image
image = np.random.randint(0,256,(6,6))
print("1-image:\n",image)

# Image Shape (Resolution)

print("\n2-Image Shape (Resolution):",image.shape)


# Cropping an Image (Slicing)
crop = image[1:4,1:4]
print("\n3-Cropping an Image (Slicing):\n",crop)

# Increase Image Brightness

bright_image = image + 50
print("\n4- Increase Image Brightness:\n",bright_image)


# Darken the Image
dark_image = image -40
print("\n5-Darken the Image:\n",dark_image)

# Normalize Pixel Values
print("\n6-Normalize Pixel Values:\n",image / 255) # we convert pixels to 0–1 range.

# Image Statistics
print("\n7-Image Statistics")
print("i-Max pixel:", np.max(image))
print("ii-Min pixel:", np.min(image))
print("iii-Average pixel:", np.mean(image))

# Edge Detection
edges = image[:,1:] - image[:,:-1]
# : → select all rows
# 1: → select columns starting from index 1
#:-1 → select all columns except the last
print("\n8-Edge Detection:\n",edges)

# Show image
plt.imshow(image,cmap="gray")
plt.title("Original Image")
plt.show()
```
